```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load a color image
image = cv2.imread('C:\\Users\\chari\\OneDrive\\Documents\\Sliit\\Y3S1\\DIP\\Lab 5\\bird.png')
gray_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

gray_3 = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
gray_3 = cv2.normalize(image,  None, 0, 2**-3, cv2.NORM_MINMAX)

gray_5 = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
gray_5 = cv2.normalize(image,  None, 0, 2**5, cv2.NORM_MINMAX)

img_size = gray_img.size

histogram = cv2.calcHist([gray_img], [0], None, [256], [0, 256])
total_pixels = img_size[0] * img_size[1]
probability = histogram / total_pixels


plt.imshow(gray_3, cmap="gray")
plt.title('Gray Intens 3')
plt.show()

plt.imshow(gray_5, cmap="gray")
plt.title('Gray Intens 5')
plt.show()

plt.imshow(gray_img, cmap="gray")
plt.title('Gray_Image')
plt.show()

print('c. Probability of occurrence of each intensity level:')
for i, prob in enumerate(probability):
    print(f'   Intensity {i}: {prob[0] / total_pixels * 100:.2f}%')

cv2.waitKey(0)
```
