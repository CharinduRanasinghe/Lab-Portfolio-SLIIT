## Addition of two Matrices

Implement elementwise Matric addition without using any functions.

```python
import numpy as np

# Define matrices A and B
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

C = np.zeros_like(A)

# Perform matrix addition manually
C = A + B


print("Matrix A:\n", A)
print("Matrix B:\n", B)
print("Matrix C (A + B):\n", C)
```

## Dot Product

```python
import numpy as np

# Define matrices A and B
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])


C = np.zeros((A.shape[0], B.shape[1]))

# Perform matrix multiplication (dot product) manually
C = np.dot(A, B)

print("Matrix A:\n", A)
print("Matrix B:\n", B)
print("Matrix C (A * B):\n", C)
```

## Cross Product

```python
import numpy as np

# Define vectors u and v
u = np.array([1, 2, 3])
v = np.array([4, 5, 6])


cross_product = np.zeros(3)

# Perform vector cross product manually
cross_product = np.zeros(3)
cross_product[0] = u[1] * v[2] - u[2] * v[1]
cross_product[1] = u[2] * v[0] - u[0] * v[2]
cross_product[2] = u[0] * v[1] - u[1] * v[0]

print("Vector u:\n", u)
print("Vector v:\n", v)
print("Vector Cross Product (u x v):\n", cross_product)

```

## Gray Scale Convertion

The values 0.2989, 0.5870, and 0.1140 are the standard luminance weights used in the RGB-to-grayscale conversion formula. These weights are based on the relative sensitivity of the human eye's three primary color receptors (red, green, and blue) to different wavelengths of light. The values are chosen to approximate the perceived brightness of each color channel.

The formula gray_value = (r * 0.2989 + g * 0.5870 + b * 0.1140) calculates the weighted sum of the red, green, and blue components of a pixel to produce a single grayscale value that represents the perceived brightness of the pixel. The sum of the weights is approximately 1, ensuring that the resulting grayscale value is within the valid range of 0 to 255.

Step 1. Load the Image
Step 2. Convert each pixel to a grayscale value using the formula above
Step 3. display the new Image

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load a color image
image = cv2.imread('C:\\Users\\chari\\OneDrive\\Documents\\Sliit\\Y3S1\\DIP\\Lab 1\\test.jpg')
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)


gray_image = np.zeros((image_rgb.shape[0], image_rgb.shape[1]), dtype=np.uint8)

# Convert image to grayscale using multiplications and divisions
for i in range(image_rgb.shape[0]):
    for j in range(image_rgb.shape[1]):
        gray_image[i, j] = int(0.299 * image_rgb[i, j, 0] + 0.587 * image_rgb[i, j, 1] + 0.114 * image_rgb[i, j, 2])



plt.imshow(gray_image, cmap='gray')
plt.title('Grayscale Image')
plt.show()
```

## Scaling Images using Matrix Manipulations

Suppose we have the following 3x3 image:



Original Image:
| 1  2  3 |
| 4  5  6 |
| 7  8  9 |

We want to scale this image by a factor of 2 in both dimensions.

Step 1: Load the original image (not shown here as it's a simple example).

Step 2: Determine the scaling factors.

    scale_x = 2 (scaling horizontally by a factor of 2)
    scale_y = 2 (scaling vertically by a factor of 2)

Step 3: Calculate new dimensions.

    The original image is 3x3.
    The new dimensions will be new_height = 3 * scale_y = 6 and new_width = 3 * scale_x = 6.

Step 4: Create an empty scaled image with new dimensions.

Step 5: Populate the scaled image using nearest-neighbor interpolation.

For each pixel (x, y) in the scaled image:

    For the pixel (0, 0) in the scaled image:
        Calculate the corresponding pixel in the original image:
            orig_x = 0 / 2 = 0
            orig_y = 0 / 2 = 0
        The nearest pixel in the original image is (1, 1), so the value in the scaled image becomes 5.

    Repeat this process for all pixels in the scaled image.

The scaled image becomes:


Scaled Image:
| 5  5  6  6  3  3 |
| 5  5  6  6  3  3 |
| 8  8  9  9  6  6 |
| 8  8  9  9  6  6 |
| 7  7  8  8  4  4 |
| 7  7  8  8  4  4 |

Step 6: Display the original image and the scaled image.

Implement this algorithm in python

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load an image
image = cv2.imread('C:\\Users\\chari\\OneDrive\\Documents\\Sliit\\Y3S1\\DIP\\Lab 1\\test.jpg')
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Define scaling factors
scale_x = 0.5
scale_y = 0.5

# Calculate new dimensions
new_height = int(image_rgb.shape[0] * scale_y)
new_width = int(image_rgb.shape[1] * scale_x)

# Initialize scaled image
scaled_image = np.zeros((new_height, new_width, 3), dtype=np.uint8)
# Scale the Pixels
for i in range(new_height):
    for j in range(new_width):
        original_i = int(i / scale_y)
        original_j = int(j / scale_x)
        scaled_image[i, j] = image_rgb[original_i, original_j]

plt.imshow(scaled_image)
plt.title('Scaled Image')
plt.show()
```

## Image Rotation

Image rotation involves rotating the entire image around a specified center by a given angle. This can be achieved using rotation matrices to transform each pixel's coordinates in the original image to its corresponding coordinates in the rotated image.

Steps to Follow:

    Step 1. Load the original image.
    Step 2. Define the rotation angle (in degrees) and the rotation center (usually the center of the image).
    Step 3. Convert the rotation angle to radians.
    Step 4. Calculate the rotation matrix based on the rotation angle.
    Step 5. Initialize an empty rotated image.
    Step 6. For each pixel in the rotated image:
                Calculate the corresponding pixel position in the original image using the rotation matrix.
                Interpolate pixel color value from the original image to the rotated image using bilinear interpolation.
    Step 7. Display the rotated image.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load an image
image = cv2.imread('C:\\Users\\chari\\OneDrive\\Documents\\Sliit\\Y3S1\\DIP\\Lab 1\\test.jpg')
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Define rotation angle (45 degrees)
theta = np.radians(45)
cos_theta = np.cos(theta)
sin_theta = np.sin(theta)

# Rotation matrix
rotation_matrix = np.array([[cos_theta, -sin_theta],
                            [sin_theta, cos_theta]])

# Initialize rotated image
rotated_image = np.zeros_like(image_rgb)

rows, cols, _ = image_rgb.shape
center_x = cols / 2
center_y = rows / 2

#Rotate the pixes to new locations
for y in range(rows):
    for x in range(cols):
        new_x = int((x - center_x) * cos_theta - (y - center_y) * sin_theta + center_x)
        new_y = int((x - center_x) * sin_theta + (y - center_y) * cos_theta + center_y)
        
        if 0 <= new_x < cols and 0 <= new_y < rows:
            rotated_image[new_y, new_x, :] = image_rgb[y, x, :]


plt.imshow(rotated_image)
plt.title('Rotated Image')
plt.show()
```

## Image Thresholding
Image thresholding is used to convert a grayscale image into a binary image by segmenting pixels into two classes based on a threshold value.

Steps to Follow:

    1. Load the original grayscale image.
    2. Define a threshold value (between 0 and 255).
    3. Create a binary image where pixels above the threshold become white (255) and pixels below or equal to the threshold become black (0).
    4. Display the original and thresholded binary images.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load a grayscale image
gray_image = cv2.imread('C:\\Users\\chari\\OneDrive\\Documents\\Sliit\\Y3S1\\DIP\\Lab 1\\test.jpg', cv2.IMREAD_GRAYSCALE)

# Define a threshold value
threshold_value = 128

rows, cols = gray_image.shape
binary_image = np.zeros_like(gray_image)

# Create a binary image using thresholding with loops and conditional statements
for y in range(rows):
    for x in range(cols):
        if gray_image[y, x] >= threshold_value:
            binary_image[y, x] = 255

# Display original and binary images
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(gray_image, cmap='gray')
plt.title('Original Grayscale Image')
plt.subplot(1, 2, 2)
plt.imshow(binary_image, cmap='gray')
plt.title('Thresholded Binary Image')
plt.show()

```

## Image Inversion

Image inversion involves reversing the intensity values of an image. Bright pixels become dark, and dark pixels become bright.

Steps to Follow:

    1. Load the original grayscale image.
    2. For each pixel, calculate the inverted intensity value by subtracting it from the maximum intensity value (usually 255).
    3. Create an inverted image using the calculated intensity values.
    4. Display the original and inverted images.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load a grayscale image
gray_image = cv2.imread('C:\\Users\\chari\\OneDrive\\Documents\\Sliit\\Y3S1\\DIP\\Lab 1\\test.jpg', cv2.IMREAD_GRAYSCALE)

# Calculate the maximum intensity value (8-bit grayscale image)
max_intensity = 255

# Calculate inverted intensity values
inverted_image = max_intensity - gray_image

# Display original and inverted images
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(gray_image, cmap='gray')
plt.title('Original Grayscale Image')
plt.subplot(1, 2, 2)
plt.imshow(inverted_image, cmap='gray')
plt.title('Inverted Image')
plt.show()
```
